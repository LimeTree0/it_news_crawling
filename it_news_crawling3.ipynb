```python
data = []
okt=Okt()
for vector in df['내용']:
    vec = okt.nouns(vector)
    #한 글자 단어 삭제
    filtered_nouns_str = [n for n in vec if len(n) > 1]
    
    #불용어 제외한 단어들로 구성된 리스트 생성
    vec = [word for word in filtered_nouns_str if word not in stop_words]
    
    data.append(vec)
```

```python
#TF-IDF 계산을 위해 명사들의 문자열 구성
doc_noun = []
for i in range(0, len(data)):
    doc_noun.append(' '.join(data[i])) #각 문서의 명사들을 str로 연결
```

```python
#텍스트 문서 모음을 단어 tf-idf 행렬로 변환
from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer(min_df = 0.01, max_df = 0.95)
#문서의 1%~95%로 나타나는 단어들을 고려
#min_df: 특정 단어가 나타나는 '문서의 수의 초소 빈도값을 설정 / 소수부분만있는 형태면 %'
X = vec.fit_transform(doc_noun)
```

```python
print(X)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
```

```python
def elbow(X):
    sse = []
    
    for i in range(1, 10):
        km = KMeans(n_clusters=i, algorithm = 'auto', random_state = 0)
        km.fit(X)
        sse.append(km.inertia_)
        print(i)
        
    plt.plot(range(1, 10), sse, marker = 'o')
    plt.xlabel('K')
    plt.ylabel("SSE")
    plt.xticks(range(1, 10))

    plt.show()
```

```python
elbow(X)
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 9, algorithm = 'auto', random_state = 0)
model.fit(X)
```

```python
print("Top terms per cluster: ")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vec.get_feature_names_out()

for i in range(9):
#   각 토픽별 30개의 메인 단어 출력
    print("Cluster %d:" % i)
    for ind in order_centroids[i, :30]:
        print('%s' % terms[ind])
    print('\n')    
```

```python
noun_dtm = [noun_dic.doc2bow(text) for text in data]
ldamodel=Lda(noun_dtm, num_topics=9, id2word=noun_dic, passes=15,random_state=0)

topics = []
for i in range(len(noun_dtm)):
    prop_sort=[]
    topic_sort=[]
    for topic , prop in ldamodel.get_document_topics(noun_dtm)[i]:  #각 문서마다 토픽 별로 해당할 확률이 부여됨
        prop_sort.append(prop)
        topic_sort.append(topic)
    topics.append(topic_sort[prop_sort.index(max(prop_sort))])  #확률 중 가장 높은 확률의 토픽을 해당 문서의 토픽으로 지정

df['topic'] = topics #모든 문서의 토픽을 데이터프레임에 추가
```

```python
#draw_wordCloud로 옮김
content_str = doc

okt = Okt()
nouns_topic = okt.nouns(content_str)
filtered_nouns_str = [n for n in nouns_topic if len(n) > 1]

#불용어 제외한 단어들로 구성된 리스트 생성
filtered_words = [word for word in filtered_nouns_str if word not in stop_words]
```

```python
match_words = [x for x in filtered_words if x in list(NRC['Korean'])] # 사전과 매칭
match_words
```

```python
emotion = []
for i in match_words:
    for emo in ['anger', 'anticipation', 'disgust', 'fear', 'joy', 'negative', 'positive', 'sadness', 'surprise', 'trust']:
        e = NRC[NRC['Korean'] == i][emo]
        if int(e) == 1:
            emotion.append(emo)
```

```python
sentiment_result1 = pd.Series(emotion).value_counts()
print(sentiment_result1, sentiment_result1.plot.bar())
```
